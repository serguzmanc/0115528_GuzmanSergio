```python
#%% PUNTO 7
num = (int)(input('Escriba un numero entero: ')) #Se pide el numero entero
num=(str)(abs(num)) #Se convierte el numero en string para poder descomponerlo
a=[] 
for i in range(len(num)):    
    a.append(num[i]) #Se introduce el digito(realmente una letra) en el vector
    a[i]=int(a[i]) #Se convierte el digito a numero      
c=[]
#Aqui se ponen en un vector los numeros que se repiten, todas las veces 
#que se repiten
for i in range(0,len(a)):
    for j in range(i+1,len(a)):
        if a[i]==a[j]:  
            c.append(a[i])             
#Aqui ya se elimitan los elementos repetidos de c, el segundo numero se 
#pone como una a , para que despues todas las a se puedan eliminar
for i in range(0,len(c)):
    for j in range(i+1,len(c)):  
        if c[i]==c[j]:
            c[j]='a'
d=[]
#Aqui se eliminan todas las a para que el nuevo vector (d) sea el ordenado
for i in range(0,len(c)):
    if c[i]!='a':
        d.append(c[i])
if len(d)!=0:        
    print('Los numeros repetidos son: ',d)   
else:
    print('No hay numeros repetidos.')
```

```python
#%% PUNTO 9
import math
x1 = (int)(input('(Primer Circulo)Escriba la coordenada x: '))
y1 = (int)(input('(Primer Circulo)Escriba la coordenada y: '))
r1 = (int)(input('(Primer Circulo)Escriba el radio: '))

x2 = (int)(input('(Segundo Circulo)Escriba la coordenada x: '))
y2 = (int)(input('(Segundo Circulo)Escriba la coordenada y: '))
r2 = (int)(input('(Segundo Circulo)Escriba el radio: '))

a = (int)(input('Coordenada a: '))
b = (int)(input('Coordenada b: '))

#Se obtiene la distancia desde la coordenada al centro, para ambos circulos
dist1= math.sqrt((x1-a)**2+(y1-b)**2)
dist2= math.sqrt((x2-a)**2+(y2-b)**2)

cont=0
#Las condiciones se basan en que si la distancia es menor que el radio
#entonces esta dentro del circulo
if (dist1<=r1):
    print('(a,b) esta dentro del circulo 1')
    cont=cont+1
else:
    print('(a,b) no esta dentro del circulo 1')    
if (dist2<=r2):  
    print('(a,b) esta dentro del circulo 2')
    cont=cont+1
else:
    print('(a,b) no esta dentro del circulo 2')
#Si entró a los 2 if, entonces quiere decir que esta dentro de los 2 circulos      
if (cont==2):
    print('(a,b) esta dentro de los dos circulos')
#Si no entró a ninguno de los 2 if, entonces cont=0
if (cont==0):
    print('(a,b) no esta dentro de los dos circulos')
```

```python
#%%PUNTO 18
tamano= (int)(input('De que tamaño quiere que sea la matriz cuadrada(n*n): '))
A=[]  #Defino la matriz
#Se empieza a pedir los valores para llenar la matriz
for i in range(tamano):
    A.append([])
    for j in range(tamano):
        A[i].append((float)(input('Introduzca un numero para la posicion '+str(i+1)+','+str(j+1)+':')))
T=[]    #Defino la matriz triangular 
Cant1=[]    #Vector donde se guardara la cantidad de unos en cada fila
#Se empieza a llenar la matriz superior. Las posiciones de los elementos en una
#matriz superior estan ubicados cuando la fila es menor o igual que la columna
for i in range(tamano):
    T.append([])
    for j in range(tamano):
        if i>j:            
            T[i].append(0)
        else:
            T[i].append(A[i][j])
    #Esto sirve para contar la cantidad de unos que hay en cada fila
    Cant1.append(T[i].count(1)) 
b=Cant1.index(max(Cant1))   #Sirve para saber que fila tiene mas unos
a=Cant1.count(max(Cant1))   #Sirve para saber cuantos maximos hay
if max(Cant1)==0:
    print('No se encuentran unos en la matriz')    
elif a==1:    
    print('El numero mayor de unos se encuentra en la fila ',b+1)
else:   #Si a!= entonces quiere decir que hay varias filas con el mismo maximo
    B=[]  #Aqui se guardaran las posiciones del vector Cant1 donde hay maximos
    for i in range(len(Cant1)):
        if Cant1[i]==max(Cant1):
            B.append(i)                        
    print('Las filas con mayor numero de unos son las siguientes: ',B)
```
